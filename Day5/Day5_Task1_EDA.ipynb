8. Plot for following analysis

   A. Country wise count of nuclear explosion

````python
by_country = df.groupby('Weapon_Source').size()

by_country.plot(kind = 'bar')
plt.title('Country Wise Count of Nuclear Explosion')
plt.xlabel('Country')
plt.ylabel('Count of Explosions')
````

B. Deployment location count

````python
deployment_loc = df.groupby('Weapon_Deployment').size()

deployment_loc.plot(kind = 'bar', figsize=(30,10))
plt.title('Deployment Location Count of Nuclear Explosion')
plt.xlabel('Deployment')
plt.ylabel('Count of Explosions')
````

````python
deployment_loc
````

C. Year Wise Count

````python
year_count = df.groupby('Year').size()
year_count.plot()
plt.title('Year-wise Count of Nuclear Explosion')
plt.xlabel('Year')
plt.ylabel('Count of Explosions')
````

D. Country wise average explosion yield

````python
yield_by_country = df.groupby('Weapon_Source')[['Yield_Lower', 'Yield_Upper']].mean()
yield_by_country

yield_by_country.plot()
plt.title('Country Wise Average Explosion Yield')
plt.xlabel('Country')
plt.ylabel('Yield')
````

E. Surface Type count

````python
count_surface_type = df.groupby('Type').size()
count_surface_type.plot(kind = 'bar')
plt.title('Surface Type Count')
plt.xlabel('Surface Type')
plt.ylabel('Count')
````

F. Surface type average explosion yield

````python
surface_type_yield = df.groupby('Type')[['Yield_Lower', 'Yield_Upper']].mean()
surface_type_yield.plot(kind = 'bar')
````

G. Purpose Distribution (Frequency)

````python
purpose = df.groupby('Purpose').size()
purpose.plot(kind = 'bar')
plt.title('Purpose Distribution')
plt.xlabel('Purpose')
plt.ylabel('Count')
````

H. Correlation between yield and depth of nuclear test

````python
corr_matrix = df[['Yield_Lower','Yield_Upper','Depth']].corr()
sns.heatmap(corr_matrix)
````

 I. Box plot of all numeric variables

````python
numeric_column_data = df[['Location.Cordinates.Latitude','Location.Cordinates.Longitude','Magnitude_Body',
'Magnitude_surface',
'Depth',
'Yield_Lower',
'Yield_Upper']]

plt.figure(figsize=(20,10))
sns.boxplot(data = numeric_column_data, orient = 'h', palette='Set3')
````
