```python
import pandas as pd
import os, io
import numpy as np
from pandas import Series, DataFrame, read_table
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
df_books = pd.read_csv("Books.csv//Books.csv")
df_books
```

```python
df_ratings = pd.read_csv("Ratings.csv//Ratings.csv")
df_ratings
```

```python
df_users = pd.read_csv("Users.csv//Users.csv")
df_users
```

```python
merged_df = pd.merge(df_books, df_ratings, on='ISBN')
new_data = pd.merge(merged_df, df_users, on = 'User-ID')
new_data
```

```python
def popularity(train,title,ids):
    #group the movies by its name and count the user frequency
    train_data_grouped = train.groupby([title])[ids].count().reset_index()  #user_id  #movie title
    # Rename the frequency (count) as score
    train_data_grouped.rename(columns = {ids: 'score'},inplace=True)
    # Sort the dataframe by score and title in the ascending order
    train_data_sort = train_data_grouped.sort_values(['score',title], ascending = [0,1])
    # Generate a new column called as rank, against score (highest score gets the lowest rank)
    train_data_sort['Rank'] = train_data_sort['score'].rank(ascending=0, method='first')
    # Filter the top 10 items
    popularity_recommendations = train_data_sort.head(10) 
    
    #return the top 10 items
    return popularity_recommendations
```

```python
popularity(new_data,'Book-Title','User-ID')
```
