Create a dataframe from a dictionary where the keys are the names of countries, the values are the gdp and indices are last 10 years. 
Now try to find the average gdp combined for all countries over past 10 years. 

```python
import pandas as pd 
import numpy as np
```

```python
data_GDP = {}

Countries = ['Germany', 'India', 'USA', 'China', 'UK', 'Norway', 'Australia', 'France', 'Italy', 'Japan']
years = [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010]

for year in years: 
    data_GDP[year] = {}
    for country in Countries: 
        data_GDP[year][country] = np.random.randint(1000000)

df = pd.DataFrame.from_dict(data_GDP, orient='index')
df
```

```python
avg_df = df.mean(axis = 1)
print(avg_df)
```

```python
avg_df_combined = avg_df.mean()
avg_df_combined
```
