# GROCERY PURCHASE APPLICATION

Let us build some application

Assume you are asked to build an inventory system of grocery items at a grocery store. 

Assuming your store can have some 20 unique items with unique ID's create a dictionary of Item ID vs item details such as MRP, Quantity, Size etc

At any time we need to keep a count of the items in the store. 

Build a menu where you can show different items along with its details. Allow the user
to purchase any item and place an order. 

Once the user places an order ensure that you have adequate 
quantity of items and if yes then create a bill of purchased items and write them to a file against an order ID. Provide him another chance to purchase the remaining items until he exits

```python
import random 
import datetime

grocery_items = {'001':['Apple',99, 100,5], '002':['Coco oil',34, 80, 2], '003':['Almonds',40, 10, 2], '004':['Oranges',56, 20, 1], '005':['ScrotchBrite',80, 24, 8], 
                 '006':['Mangoes',100,2, 9], '007':['Cheese',23, 90, 2], '008':['Brown Bread',90,1,2], '009':['Ramen',29, 7,1], '010':['Spice Mix',38, 4, 3],
                 '011':['Rice',59, 30, 4], '012': ['Lentils',93, 10, 1], '013':['Sugar',63, 20, 1], '014':['Salt',80,45, 7], '015':['Honey',47, 8, 9],
                 '016':['Snack Bar',50, 8, 6], '017':['Tea',49, 9, 1], '018':['Coffee',21, 10, 8], '019':['Milk',90,7,8], '020':['Dishwasher',10, 90, 8] }


def purchase(inventory):
    
    order_id = 0
    
    while (True):
        print ('\033[1m' + 'MENU'+'\033[0m')
        print("ItemID - Name - MRP - Quantity Available")

        for key, values in inventory.items():
            print(key, end=' - ')
            print(inventory[key][0], end=' - ')
            print(inventory[key][1], end = ' - ')
            print(inventory[key][2])

        itemlist= dict()

        continue_key = input("Type continue to proceed. ")

        if continue_key == 'continue':
            while(True):
                print("Enter item ID you want to buy or enter Confirm to proceed: ")
                choice = input()

                if choice in inventory:
                    print("Enter the quantity you want to buy: ")
                    choice_quan = int(input())
                    
                    if inventory[choice][2] >= choice_quan:
                        
                        itemlist[choice] = choice_quan
                        inventory[choice][2] -= choice_quan
                        
                    else:
                        print("Sorry, insufficient quantity available.")

                elif choice == 'Confirm':
                    break
            print("Your order_id = ", order_id)



        def bill(itemlist1, inventory1): 

            bill = dict()
            grandtotal = 0

            for key, value in itemlist1.items():

                bill[key] = [inventory1[key][0], inventory1[key][1], itemlist1[key], itemlist1[key]*inventory[key][1]]
                grandtotal += itemlist1[key]*inventory[key][1]
                
            if order_id == 0:
                with open('mybill.txt', 'w') as file:
                    file.write("BILL \n")
                    file.write(f"Order ID: {order_id}\n")
                    for key, details in bill.items():
                        file.write(f"Item ID: {key}   Name:{details[0]}   MRP: {details[1]}  Quantity: {details[2]}  Total Price: {details[3]}\n")
                    file.write(f"Grand Total: {grandtotal}\n\n")
            else: 
                with open('mybill.txt', 'a') as file:
                    file.write(f"Order ID: {order_id}\n")
                    for key, details in bill.items():
                        file.write(f"Item ID: {key}   Name:{details[0]}   MRP: {details[1]}   Quantity: {details[2]}  Total Price: {details[3]}\n")
                    file.write(f"Grand Total: {grandtotal}\n\n")
                
            print("Do you want to see your bill? (Y/N)")
            showbill = input()
            
            if showbill == 'Y':
                print("Order ID: ", order_id)
                for key, details in bill.items():
                    
                    print(f"Item ID: {key}, Name:{details[0]}, MRP: {details[1]}, Quantity: {details[2]}, Total Price: {details[3]}\n")
                print(f"Grand Total: {grandtotal}")
                

        bill(itemlist, inventory)
        
        print("Make another purchase (Y/N):  ")
        anotherpurchase = input()
        
        if anotherpurchase == 'Y':
            order_id +=1
            continue
            
        else: 
                break 
    print("Thank you for shopping with us! We are closed now :) See you next time")
        


purchase(grocery_items)
```
